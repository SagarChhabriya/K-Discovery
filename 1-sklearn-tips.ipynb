Let's suppose you want to apply the OneHotEncoder on the Embarked column of the titanic dataset and there are 3 feature S, Q, and C. You have 
splitted the train and test data. You apply fit on the training data X_train the ohe object will learn the parameters (S,Q, and C) and now you can apply the transform method to transform the X_train data and X_test data without explicitly applying the fit method on test data to prevent data leakage.

Why fit only once on the X_train and not on the X_test?
let's suppose after the train_test_split the Embarked column of the X_train set has S,Q, and E but the X_test column has only **Q**. You applied the fit on the X_train and there were present all the S Q E and later you applied the transform on the both X_train and X_test so in both X the value of Q will be the 0 1 0. But you apply the fit on the X_train separately (Q = 0 1 0 )  and fit on the X_test separately which results in the (Q = 1) becuase there were no other values such as S or E. **That's why you apply the fit method only once.**
